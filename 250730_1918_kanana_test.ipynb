```python
import time
import sys
import logging

# ë¡œê¹… ì„¤ì •
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Colab í™˜ê²½ì— ë§ê²Œ ê²½ë¡œ ìˆ˜ì • ---
# 1. í”„ë¡œì íŠ¸ ê²½ë¡œ ì„¤ì •
# sys.path.append("/content/lily_generate_package")
sys.path.append("/content")

from lily_llm_api.models import get_model_profile

def run_test():
    """ë…ë¦½ì ì¸ ëª¨ë¸ ìƒì„± í…ŒìŠ¤íŠ¸ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤."""

    model_id = "kanana-1.5-v-3b-instruct"
    prompt = "ì•ˆë…•í•˜ì„¸ìš”! ìê¸°ì†Œê°œë¥¼ í•´ì£¼ì„¸ìš”."

    logging.info(f"'{model_id}' ëª¨ë¸ í”„ë¡œí•„ ë¡œë”© ì‹œì‘...")
    profile = get_model_profile(model_id)

    # 2. í”„ë¡œí•„ ë‚´ë¶€ì˜ ê²½ë¡œë„ Colabì— ë§ê²Œ ìˆ˜ì •
    profile.local_path = "/content/lily_generate_package/lily_llm_core/models/kanana-1.5-v-3b-instruct"

    # ì´ì œ ë¡œë”©ê³¼ ìƒì„±ì„ ì‹œì‘í•©ë‹ˆë‹¤.
    model, tokenizer = profile.load_model()
    logging.info("âœ… ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì € ë¡œë”© ì„±ê³µ!")

    # (ì´í•˜ ì½”ë“œëŠ” test_generation.pyì™€ ë™ì¼í•©ë‹ˆë‹¤)
    logging.info("í”„ë¡¬í”„íŠ¸ í¬ë§·íŒ… ì¤‘...")
    messages = [{"role": "user", "content": prompt}]
    formatted_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(formatted_prompt, return_tensors="pt").to(model.device)

    gen_config = profile.get_generation_config()
    gen_config['max_new_tokens'] = 50 # ë‹µë³€ì„ ì¡°ê¸ˆ ë” ê¸¸ê²Œ ë³´ë„ë¡ ì„¤ì •

    logging.info(f"--- í…ìŠ¤íŠ¸ ìƒì„± ì‹œì‘ (ìµœëŒ€ {gen_config['max_new_tokens']} í† í°) ---")
    start_time = time.time()

    outputs = model.generate(**inputs, **gen_config)

    end_time = time.time()
    duration = end_time - start_time

    logging.info(f"--- í…ìŠ¤íŠ¸ ìƒì„± ì™„ë£Œ! (ì†Œìš” ì‹œê°„: {duration:.2f}ì´ˆ) ---")

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    clean_response = profile.extract_response(response)

    print("\n" + "="*50)
    print(f"ğŸ“ í”„ë¡¬í”„íŠ¸: {prompt}")
    print(f"ğŸ¤– ì •ì œëœ ì‘ë‹µ:\n{clean_response}")
    print("="*50)

# ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰
run_test()
```
